```python
import matplotlib.pyplot as plt
import numpy
```

Creating a random number from the range 0 to 5 of 250 values

```python
randomvalue = numpy.random.uniform(0.0, 0.5, 1000)
plt.hist(randomvalue,5)
plt.show()
```

```python
# Random value to normalise
randomvalue = numpy.random.normal(0.0, 5.0, 1000)
plt.hist(randomvalue,500)
plt.show()
```
